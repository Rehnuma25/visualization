# Bar

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
plt.figure(figsize=(8,5))
colors = ["red", "green", "blue"]
plt.barh(fruits, sales,color= colors)  #use brah not bar

plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
x = ["Red", "Blue", "Pink", "Orange", "Indigo", "Yellow", "Black", "White", "chayan", "gay", "Mejenda", "chocolet"]
y = ["10", "50", "70", "40", "30", "50", "10", "60", "63", "48", "12", "22"]
y = list(map(int, y)) # Convert y to integers (if we use seaborn that's not need )
colors = ["red", "blue", "pink", "orange", "indigo", "yellow", "black", "gray", "cyan", "purple", "brown"] #12 valid colors

plt.figure(figsize=(8,7))
plt.bar(x, y, color=colors)
plt.title("Colours Range")
plt.xlabel("Colour")
plt.ylabel("Range")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
x = [1,5,4,7,5,7,7]
y = [10,20,30,25,40,35,10]
plt.figure(figsize=(5,6))
sns.barplot(x=x,y=y, color = "blue" )
plt.xlabel("value of x")
plt.ylabel("value of y")
plt.show()
```

# pie chart

```python
x = ["bike", "car", "bus","Auto"]
y = [40, 50, 60,30]

fig, axs = plt.subplots(1, 2, figsize=(8, 6))  # 1 row, 2 columns

#1st pie chart
axs[0].pie(y, labels=x)
axs[0].set_title("Basic Pie Chart")

# 2nd pie chart
axs[1].pie(y, labels=x, startangle=160, autopct="%1.2f%%")
axs[1].set_title("Stylish Pie Chart")
```

```python
y = [35, 25, 25, 15]
x = ["Apples", "Bananas", "Cherries", "Dates"]
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

#1st pie chart
axs[0].pie(y, labels=x, startangle=160, autopct="%1.2f%%")
axs[0].set_title("Basic Pie Chart")# 1 row, 2 columns

#2nd pie chart
myexplode=[0.2,0.3,0,0] # no other name use it cannot work.
plt.pie(y, labels =x, explode = myexplode, startangle=20,  autopct="%1.2f%%" ,shadow = True)
plt.show() 
```

# boxplot

```python
data = [1,10,50,100,190,80,70,160,150,70,180,60,-160,-600,700,600]
plt.figure(figsize=(6,5))
sns.boxplot(data, color = "g", linewidth=2, linecolor="red")
plt.title("boxplot")
plt.xlabel("value of x")
plt.ylabel("value of y")
plt.show()
```

# count plot

```python
clas = ["a", "b", "c", "b", "c", "c", "d"]
huy  = ["x", "x", "y", "x", "y", "y", "z"]  # same length as clas
plt.figure(figsize=(5, 4))
sns.countplot(x=clas, hue=huy, palette="Set1") # set1 just use for color
plt.xlabel("value of x")
plt.title("Count Plot")
plt.show()
```

# pair plot

```python
df = pd.DataFrame({
    "x": [1,2,3,4,5,6],
    "y": [22,3,4,5,6,7],
    "z": [3,4,5,6,7,8],
    "class": ["a", "b", "c", "b", "c", "c"]
})
sns.pairplot(df, hue="class") # Pairplot with hue
plt.show()
```
