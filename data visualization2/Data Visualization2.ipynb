# stack area chart

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 2, 3, 3, 4]
y3 = [1, 1, 1, 1, 1]

plt.stackplot(x, y1, y2, y3, labels=['1', '2', '3'], colors=['#f4a261','#2a9d8f','#e76f51'])
plt.legend(loc='upper left')
plt.title("Stacked Area-Chart")
plt.show()
```

# Candlestick chart

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Candlestick(
    x=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    open=[33, 34, 32, 36, 35],
    high=[35, 36, 34, 38, 37],
    low=[32, 33, 31, 34, 34],
    close=[34, 35, 33, 37, 36]
)])
fig.update_layout(title='Candlestick Chart')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.arange(1, 6)
y = np.array([3, 5, 4, 6, 7])
df = pd.DataFrame({'x': x, 'y': y})

sns.barplot(x='x', y='y', data=df, color='#ff8000')
sns.regplot(x='x', y='y', data=df, scatter=False, color='blue')
plt.title("Bar Chart with Trendline")
plt.show()
```

# Donut chart

```python
sizes = [30, 25, 25, 20]
labels = ['part1', 'part2', 'part3', 'part4']
colors = ['#ff0000','#0040ff','#ff8000','#40ff00']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, wedgeprops={'width':0.5})
ax.set(aspect="equal", title='Donut Chart')
plt.show()
```

# correlation heat map

```python
# Sample dataset
data = {
    'Math': [80, 85, 78, 92, 88,33,90,67,45,66],
    'Physics': [82, 79, 74, 91, 86,77,88,99,100,54],
    'Chemistry': [78, 81, 80, 89, 12,34,55,67,89,100],
    'Biology': [70, 75, 72, 85, 80,67,89,90,77,88]
}

df = pd.DataFrame(data)

## Calculate correlation matrix
corr_matrix = df.corr()
#create a map
plt.figucorr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.7)
plt.title("Correlation-map")
plt.show()
```

# Correlation Heatmap with load dataset

```python
df = pd.read_csv('online_classroom_data.csv', dtype=str)

# Replace comma with dot and convert to numeric where possible
for col in df.columns:
    df[col] = df[col].str.replace(',', '.')
    df[col] = pd.to_numeric(df[col], errors='coerce')  # invalid values become NaN

# Drop columns that are entirely NaN
df = df.dropna(axis=1, how='all')

# Compute correlation only on numeric columns
corr_matrix = df.corr()

#  heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.7)
plt.title("Correlation-map")
plt.show()
```
